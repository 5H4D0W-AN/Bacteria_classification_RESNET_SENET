```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
X = np.load('Data/X_reference.npy')
y = np.load('Data/y_reference.npy')
class_num=0
num_each=2000
```

```python
print(X)
print(y.shape)
print(y[2000],y[1999],y[2001],y[1998])
print(y)

yhot=[]
```

```python
#data generation for 1 bacteria and 1 foreign
x_gen = []
y_lbl = []
for i in range(4):
    #with_foreign
    for j in range(i*2000, (i+1)*2000):
        foreign = np.random.randint(8000, 16000)
        cf=random.uniform(0.0,1)
        c1=1-cf
        x = c1*X[j]+cf*X[foreign]
        x_gen.append(x)
        yyy=[0.0]*5
        yyy[i]=c1
        yyy[4]=cf
        yhotenc=[0.0]*5
        yhotenc[i]=1.0
        yhotenc[4]=1.0
        yhot.append(yhotenc)
        y_lbl.append(yyy+[class_num])
        
    class_num+=1
    #without_foreign
    for j in range(i*2000, (i+1)*2000):
        x = X[j]
        x_gen.append(x)
        yyy=[0.0]*5
        yyy[i]=c1
        yhotenc=[0.0]*5
        yhotenc[i]=1.0
        yhot.append(yhotenc)
        y_lbl.append(yyy+[class_num])
    class_num+=1
xx = np.array(x_gen)
yy = np.array(y_lbl)
yyc=[]
print(yhot)
for i in yy:
    yyc+=[i[-1]]
np.save('Data/y_datagen_for1_withforeign_class', yyc)
np.save('Data/X_datagen_for1_withforeign', xx)
np.save('Data/y_datagen_for1_withforeign', yy)
```

```python
#data generation for 2 bacteria and 1 foreign


from itertools import permutations
l = list(permutations(range(0, 4)))

x_gen = []
y_lbl = []
for i in range(4):
    for j in range(i+1,4):
        #with_foreign
        for _ in range(num_each):
            foreign = np.random.randint(8000, 16000)
            first = np.random.randint(i*2000, (i+1)*2000)
            second = np.random.randint(j*2000, (j+1)*2000)
            cf=random.uniform(0.0,1)
            c1 = random.uniform(0.0,1-cf)
            c2 = 1 - c1 -cf
            conc=[c1,c2,cf]
            random.shuffle(conc)
            c1,c2,cf=conc[0],conc[1],conc[2]
            x = c1*X[first] + c2*X[second] +cf*X[foreign]
            x_gen.append(x)
            yyy=[0.0]*5
            yyy[i]=c1
            yyy[j]=c2
            yyy[4]=cf
            yhotenc=[0.0]*5
            yhotenc[i]=1.0
            yhotenc[j]=1.0
            yhotenc[4]=1.0
            yhot.append(yhotenc)
            y_lbl.append(yyy+[class_num])
        class_num+=1
        #without_foreign
        for _ in range(num_each):
            first = np.random.randint(i*2000, (i+1)*2000)
            second = np.random.randint(j*2000, (j+1)*2000)
            c1 = random.uniform(0.0,1)
            c2 = 1 - c1
            conc=[c1,c2]
            random.shuffle(conc)
            c1,c2=conc[0],conc[1]
            x = c1*X[first] + c2*X[second]
            x_gen.append(x)
            yyy=[0.0]*5
            yyy[i]=c1
            yyy[j]=c2
            yhotenc=[0.0]*5
            yhotenc[i]=1.0
            yhotenc[j]=1.0
            yhot.append(yhotenc)
            y_lbl.append(yyy+[class_num])
        class_num+=1
xx = np.array(x_gen)
yy = np.array(y_lbl)
yyc=[]
for i in yy:
    yyc+=[i[-1]]

np.save('Data/y_datagen_for2_withforeign_class', yyc)
np.save('Data/X_datagen_for2_withforeign', xx)
np.save('Data/y_datagen_for2_withforeign', yy)
```
